# Bicycle data 2018 in Berlin

## Preparing data

```python
from pathlib import Path
import os

data_dir = str(Path.home()) + r'\tubCloud2\Master\Git\Cycling_traffic_volume_Berlin'
```

```python
import pandas as pd
# load data
counts2018 = pd.read_csv(os.path.join(data_dir, r'Bicycle_volume2018.csv'), delimiter=';', index_col=0, parse_dates=True)
```

```python
# load more data
counts2019 = pd.read_csv(os.path.join(data_dir, r'Bicycle_volume2019.csv'), delimiter=';', index_col=0, parse_dates=True)
```

```python
counts2018.head()
```

```python
# concatenate 2018 and 2019 data
counts = pd.concat([counts2018, counts2019])
#counts.tail()
```

```python
counts.shape
#counts.columns[3]
```

```python
# sum up the overall counts (on hourly basis) to daily counts
daily_counts = counts.resample('d').sum()
daily_counts.head()
```

```python
daily_counts.shape
```

```python
# Station names
station_names = ['JannowitzbrÃ¼cke', 'InvalidenstraÃŸe', 'OberbaumbrÃ¼cke', 'Frankfurter Allee', 'Berliner StraÃŸe',
                 'Schwedter Steg', 'PrinzregentenstraÃŸe', 'KlosterstraÃŸe', 'Breitenbachplatz', 'YorckstraÃŸe',
                 'MonumentenstraÃŸe', 'Mariendorfer Damm', 'Maybachufer', 'Kaisersteg', 'AlberichstraÃŸe',
                 'Paul-und-Paula-Uferweg', 'MarkstraÃŸe']
            
#start_numbers = [str(n) for n in list(range(2,27))]
start_numbers = ['02', '03', '05', '06', '10', '12', '13', '15', '17', '18', '19', '20', '21', '23', '24', '26', '27']

#counts.loc[:, counts.columns.str.startswith("02")]
```

```python
# Function that returns columns that start with the same prefix
def return_same_columns(cols):
    return [val for col in cols for val in start_numbers if col.startswith(val)]
```

```python
# sum up the counts for each counting station
daily_counts = daily_counts.groupby(return_same_columns(daily_counts.columns), axis=1).sum()
```

```python
# rename columns
daily_counts.columns = station_names
```

```python
daily_counts.tail()
```

```python
# one-hot-encoding for days of the week
# assumption 1: number of bicycle rides may depend on day of the week
# store values 1 or 0 in columns if the day of the week is equal to this day or not
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
for i in range(7):
    daily_counts[days[i]] = (daily_counts.index.dayofweek == i).astype(float)

#daily_counts.tail(10)
```

```python
# use the Berlin holiday data
# assumption 2: bicycle trips might be different on holidays
from datetime import date
import holidays

berlin_holidays = holidays.CountryHoliday('DE', prov='BE')
# use holidays only for 2018 and 2019
date(2018, 1, 1) in berlin_holidays
date(2019, 1, 1) in berlin_holidays
print(berlin_holidays.years)
```

```python
# for date, name in sorted(holidays.DE(prov='BE', years=[2018,2019]).items()):
#      print(date, name)
```

```python
# construct series object with column name 'holiday'
# contains the date for each holiday as index
holidays_series = pd.Series(1, index=berlin_holidays, name='Holiday')
holidays_series.tail()
```

```python
# append holiday column
daily_counts = daily_counts.join(holidays_series)
```

```python
# fill the NaN values in place (directly changed, not copied)
daily_counts['Holiday'].fillna(0, inplace=True)
daily_counts.tail(7)
```

```python
# assumption 3: hours of daylight affect number of bicycle rides
import numpy as np
from datetime import datetime

def hours_of_daylight(date, axis=23.44, latitude=52.52):
    """Compute hours of daylight in Berlin for the given date"""
    days = (date - datetime(2000, 12, 21)).days
    m = (1. - np.tan(np.radians(latitude))
    * np.tan(np.radians(axis) * np.cos(days * 2 * np.pi / 365.25)))
    return 24. * np.degrees(np.arccos(1 - np.clip(m, 0, 2))) / 180.

daily_counts['Daylight_hrs'] = list(map(hours_of_daylight, daily_counts.index))

#daily_counts.head()
```

```python
# assumption 4: bicycle trips are weather dependent
# source: https://meteostat.net/de/station/10382?t

weather2018 = pd.read_csv(os.path.join(data_dir, 'Weather_Berlin2018.csv'), index_col='date', delimiter=',', parse_dates=True)
weather2019 = pd.read_csv(os.path.join(data_dir, 'Weather_Berlin2019.csv'), index_col='date', delimiter=',', parse_dates=True)
weather = pd.concat([weather2018, weather2019])
```

```python
# add a colum that indicates wether it was a dry day or not
weather['Dry Day'] = (weather['prcp'] == 0).astype(int)
weather.head()
```

```python
# join bicycle and weather data
bicycle = daily_counts.join(weather)
#bicycle.tail()
```

```python
# add a 'year' column (float) that gives percentage of the days passed per year (0...1)
# -> see annual increase or decrease in daily crossing
bicycle['Year'] = (bicycle.index - bicycle.index[0]).days / 365.
#bicycle.head()
```

```python
# drop all still existing NaN values
print(bicycle.shape)
bicycle.dropna(axis=0, how='any', inplace=True)
print(bicycle.shape)
```

## Linear Regression Model

```python
from sklearn.model_selection import train_test_split
# exctract input features X and target columns y
# '*' -> unpack both iterables
input_column_names = [*days, 'Holiday', 'Daylight_hrs', 'prcp', 'Dry Day', 'tmin', 'tmax', 'Year']

X = bicycle[input_column_names]
y = bicycle['JannowitzbrÃ¼cke']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# import, train and fit model
from sklearn.linear_model import LinearRegression

                        # data is expected to be centered
                        # day p. week work as a day-specific intercept
model = LinearRegression(fit_intercept=False)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
# plot ground truth data against predicted data
bicycle['Jan_Predicted'] = model.predict(X)
bicycle[['JannowitzbrÃ¼cke', 'Jan_Predicted']].plot(figsize=(10, 8), alpha=0.5);
```

```python
# see how much each feature contributes to the daily bicycle count
params = pd.Series(model.coef_, index=X.columns)
params
```

```python
# bootstrap resampling
# to see the error of variation of each coefficient
from sklearn.utils import resample

np.random.seed(1)

err = np.std([model.fit(*resample(X, y)).coef_ for i in range(1000)], 0) 
```

```python
 print(pd.DataFrame({'effect': params.round(0),
'error': err.round(0)}))
```
